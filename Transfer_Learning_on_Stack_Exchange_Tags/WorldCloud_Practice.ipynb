```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from StopWords import stop_words
import nltk

%matplotlib inline
```

```python
topic_list = ['cooking', 'biology', 'crypto', 'diy', 'robotics', 'travel']

for topic in topic_list:
    tags = np.array(pd.read_csv('./src/' + topic + '.csv', usecols=['tags'])['tags'])
    text = ''

    for ind, tag in enumerate(tags):
        text = " ".join([text, tag])
    text.strip()

    wordcloud = WordCloud(background_color='white', width=600, height=400, max_font_size=50, max_words=40).generate(text)
    plt.imshow(wordcloud)
    plt.title('WordCloud for ' + topic)
    plt.axis("off")
    plt.show()
```

```python
titles = np.array(pd.read_csv('./src/cooking.csv', usecols=['title'])['title'])
text = ''
for title in titles:
    text = " ".join([text, title])
text.strip()
wordcloud = WordCloud(background_color='white', width=600, height=400, max_font_size=50, max_words=40).generate(text)
plt.imshow(wordcloud)
plt.title('WordCloud for cooking')
plt.axis("off")
plt.show()
```
